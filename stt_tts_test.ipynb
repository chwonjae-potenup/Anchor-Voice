````python
import os
import json
import pandas as pd
import whisper
from dotenv import load_dotenv
from openai import OpenAI


load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)
````

````python
model = whisper.load_model("base")
````

````python
audio_path = "./downloads/fss_audio/대출사기형/test.mp3"
````

````python
result = model.transcribe(audio_path, language="ko")
````

````python
print(result)
````

````python
for seg in result["segments"]:
    start = seg["start"]
    end = seg["end"]
    text = seg["text"].strip()
    print(f"{start:.2f}s ~ {end:.2f}s | {text}")
````

````python
rows = []

for seg in result["segments"]:
    rows.append({
        "start": round(seg["start"], 2),
        "end": round(seg["end"], 2),
        "text": seg["text"].strip()
    })

df_segments = pd.DataFrame(rows)
df_segments.head()
````

````python
segment_text = "\n".join(
    [f"{row.start:.2f}-{row.end:.2f}: {row.text}" for row in df_segments.itertuples()]
)

print(segment_text[:2000])
````

````python
system_prompt = """
너는 보이스피싱 통화 분석 전문가다.

주어진 STT 세그먼트 텍스트를 분석해서
아래 항목을 JSON 배열 형태로 정리하라.

각 항목은 반드시 아래 키를 가져야 한다:
- fraud_type: 수법 유형
- sub_type: 세부 분류
- summary: 수법 설명(요약)
- keyword: 주요 키워드
- evidence: 시간대와 원문 근거 목록

반드시 아래 JSON 스키마를 정확히 따르라.

절대 다른 키를 사용하지 말 것.
"text" 키만 사용하고 "sentence" 같은 다른 이름은 금지.

스키마:
{
  "fraud_type": "...",
  "sub_type": "...",
  "summary": "...",
  "keyword": [...],
  "evidence": [
    {
      "time": "...",
      "text": "..."
    }
  ]
}

FRAUD_TYPES = [
    "수사기관 사칭형",
    "금융기관 사칭형",
    "가족·지인 사칭형",
    "스미싱",
    "투자사기"
]

SUB_TYPES = {
    "수사기관 사칭형": [
        "검찰 사칭",
        "고립 유도",
        "소환장 압박"
    ],
    "금융기관 사칭형": [
        "저금리 대출",
        "대환대출 유도",
        "거래실적 요구"
    ],
    "가족·지인 사칭형": [
        "자녀 사칭",
        "폰 고장 핑계",
        "긴급 송금 요구"
    ]
}

fraud_type은 반드시 아래 중 하나만 선택하라:

- 수사기관 사칭형
- 금융기관 사칭형
- 가족·지인 사칭형
- 스미싱
- 투자사기

sub_type은 해당 fraud_type에 속하는 항목 중 하나만 선택하라.

절대 새로운 값을 만들지 말 것.

주의사항:
1. evidence에는 실제 시간대와 문장을 넣어라.
2. 없는 내용을 지어내지 말고, STT 근거가 있는 것만 써라.
3. 여러 수법이 섞여 있으면 여러 항목으로 나눠라.
4. 출력은 반드시 JSON만 반환하라.
"""
````

````python
user_prompt = f"""
다음은 보이스피싱 의심 통화의 STT 세그먼트다.

{segment_text}
"""

response = client.responses.create(
    model="gpt-4.1-mini",
    input=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ]
)

llm_text = response.output_text
print(llm_text)
````

````python
clean_text = llm_text.replace("```json", "").replace("```", "")
analysis_data = json.loads(clean_text)
analysis_data
````

````python
table_rows = []

for item in analysis_data:
    evidence_text = "\n".join(
        [f"{ev['time']} {ev['text']}" for ev in item["evidence"]]
    )

    table_rows.append({
        "수법 유형": item["fraud_type"],
        "세부 분류": item["sub_type"],
        "수법 설명(요약)": item["summary"],
        "주요키워드": item["keyword"],
        "실제 시나리오 / 원본 텍스트 전체": evidence_text
    })

df_result = pd.DataFrame(table_rows)
df_result
````

````python
df_result.to_excel("phishing_analysis_result.xlsx", index=False)
print("저장 완료")
````
